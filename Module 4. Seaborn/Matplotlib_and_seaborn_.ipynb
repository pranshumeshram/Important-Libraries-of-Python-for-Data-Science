```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='Sine Wave', color='red', linestyle='--')
plt.plot(x, y2, label='Cosine Wave', color='blue', linewidth=2)
plt.title('Sine and Cosine Waves')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

## Time Series Line Chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate time series data
dates = pd.date_range('2024-01-01', periods=100)
data = pd.DataFrame({'Date': dates, 'Value': (50 + np.random.randn(100).cumsum())})

#data.head()
```

```python
# Plot
plt.figure(figsize=(10, 5))
plt.plot(data['Date'], data['Value'], color='green')
plt.title('Time Series Example')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

## Histogram for Distribution Analysis

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
data = np.random.normal(loc=0, scale=1, size=1000)
#data
```

```python
# Plot
plt.hist(data, bins=30, edgecolor='black', color='purple')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
